## Ask a Question about the data
- Try to answer a question you have about the data using a plot or statistics

What are the locations with fastest roller coasters (Minimum 10)

```python
new_df['Location'].value_counts()
```

```python
ax2 = new_df.query('Location != "Other"') \
                    .groupby('Location')['Speed_mph'] \
                    .agg(['mean', 'count']) \
                    .query('count >= 10') \
                    .sort_values('mean')['mean'] \
                    .plot(kind='barh', 
                      figsize =(12,4),
                      title= 'Average Coast Speed by Location')
ax2.set_xlabel('Average Coaster Speed')
plt.show()    
```
